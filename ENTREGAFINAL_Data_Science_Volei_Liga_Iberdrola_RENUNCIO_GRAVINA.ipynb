```python
# Creacion de 1 columna nueva llamada "Posicion_final_Ant" donde ira un valor entre 1 al 12 para determinar en que posicion se quedaron la temporada pasada,
# Si el equipo no estaba en Superliga 1, al venir de Superliga 2 y ascender se entiende que salio primero, pero de la liga 2, por eso quedara en 13 su valor.

# Crear un índice para las temporadas
temp_dict = {temp: i for i, temp in enumerate(temp_lista)}

# Agregar la columna "Posicion_final_Ant"
def obtener_posicion_anterior(row):
    temporada_actual = row['Temp']
    equipo = row['Equipo']

    # Buscar la temporada anterior en la lista
    if temporada_actual in temp_dict and temp_dict[temporada_actual] > 0:
        temporada_anterior = temp_lista[temp_dict[temporada_actual] - 1]

        # Buscar la posición final del mismo equipo en la temporada anterior
        anterior = df_equipos[(df_equipos['Equipo'] == equipo) & (df_equipos['Temp'] == temporada_anterior)]
        if not anterior.empty:
            return anterior.iloc[0]['Posicion_final_act'] # si el equipo esta en la temporada pasada pone la posicion final de dicha temporada
    return 13  # Si no hay temporada anterior o datos previos se le pone el valor 13 porque ascendieron (estaban en superliga2)

# Aplicar la función a cada fila
df_equipos['Posicion_final_Ant'] = df_equipos.apply(obtener_posicion_anterior, axis=1)
```

**COHESION DE EQUIPO ENTRE TEMPORADAS**

# **Escala de Cohesión**:
Este proceso calcula una métrica de cohesión del equipo basada en:

**Datos base:** Combinar jugadoras por nombre, enlazar temporadas consecutivas, llenar datos faltantes.

**Continuidad**: Contar cuántas temporadas seguidas una jugadora permanece en el mismo equipo.

**Cohesión por participación (Weighted Cohesion)**: Medir cuánto del tiempo jugado se retiene en la temporada siguiente, dando más peso a jugadoras con mayor continuidad.

**Cohesión por rendimiento (Performance Cohesion)**: Similar, pero basado en métricas de desempeño.

**Cálculo final = "Cohesion"**: Combinar ambas cohesion (participación y rendimiento) con pesos establecidos (60% y 40%).

**Integración**: Unir este valor de cohesión al DataFrame de equipos; asignar cohesión fija (0.5) a equipos con cierta clasificación previa; reemplazar nulos por 0.

**Propósito**: Ayudar a entender la influencia de la continuidad y el rendimiento retenido de las jugadoras en el desempeño futuro del equipo.

```python
# Creacion de Prescriptor Cohesion:

# Paso 1: Preparar los datos
player_data = df_jugadoras[['Equipo', 'Temp', 'Nombre', 'Apellido', 'Coef_jugado',
                            'Eficacia_Ataques_punto', 'Efectividad_Saque_porcentaje',
                            'Efectividad_Bloqueo_porcentaje', 'Eficacia_Recepcion_porcentaje']].copy()

player_data['Player'] = player_data['Nombre'] + " " + player_data['Apellido']

# Crear 'Next_Temp' utilizando temp_lista
player_data['Next_Temp'] = player_data['Temp'].apply(lambda x: temp_lista[temp_lista.index(x) + 1] if x in temp_lista[:-1] else None)

# Fusionar para encontrar continuidad de jugadores
player_data = player_data.merge(
    player_data[['Equipo', 'Temp', 'Player', 'Coef_jugado', 'Eficacia_Ataques_punto',
                 'Efectividad_Saque_porcentaje', 'Efectividad_Bloqueo_porcentaje',
                 'Eficacia_Recepcion_porcentaje']],
    left_on=['Equipo', 'Next_Temp', 'Player'],
    right_on=['Equipo', 'Temp', 'Player'],
    suffixes=('', '_Next'),
    how='left'
)

# Manejar valores nulos después de la fusión
player_data = player_data.fillna(0)  # Reemplazar NaN por 0

# Crear la columna 'Temporadas_Juntas'
player_data['Temporadas_Juntas'] = 1

# Iterar sobre el DataFrame para calcular 'Temporadas_Juntas'
for index, row in player_data.iterrows():
    # Verificar si la jugadora estuvo en el mismo equipo en la temporada anterior
    previous_season = temp_lista[temp_lista.index(row['Temp']) - 1] if row['Temp'] in temp_lista[1:] else None
    if previous_season:
        previous_team = player_data[(player_data['Player'] == row['Player']) & (player_data['Temp'] == previous_season)]['Equipo'].values
        if len(previous_team) > 0 and previous_team[0] == row['Equipo']:
            # Incrementar 'Temporadas_Juntas' si estuvo en el mismo equipo
            player_data.loc[index, 'Temporadas_Juntas'] = player_data.loc[index, 'Temporadas_Juntas'] + 1

# Paso 2: Calcular la Cohesión Ponderada por Participación (con factor de temporadas juntas)
def weighted_cohesion(group):
    if group['Next_Temp'].iloc[0] in temp_lista:
        total_participation = group['Coef_jugado'].sum()
        retained_participation = group['Coef_jugado_Next'].sum()
        if total_participation > 0 and retained_participation is not None:  # Verificar si retained_participation es nulo
            cohesion = retained_participation / total_participation
            cohesion_factor = 1 + group['Temporadas_Juntas'].mean() * 0.1  # Factor de ponderación
            return cohesion * cohesion_factor
    return 0 # Devolver 0 si hay valores nulos o total_participation es 0


# Paso 3: Calcular la Cohesión Ajustada por Rendimiento
def performance_cohesion(group):
    if group['Next_Temp'].iloc[0] in temp_lista:  # Asegurar que la siguiente temporada existe en temp_lista
        total_performance = (
            group['Eficacia_Ataques_punto'] +
            group['Efectividad_Saque_porcentaje'] +
            group['Efectividad_Bloqueo_porcentaje'] +
            group['Eficacia_Recepcion_porcentaje']
        ).sum()
        retained_performance = (
            group['Eficacia_Ataques_punto_Next'] +
            group['Efectividad_Saque_porcentaje_Next'] +
            group['Efectividad_Bloqueo_porcentaje_Next'] +
            group['Eficacia_Recepcion_porcentaje_Next']
        ).sum()
        if total_performance > 0 and retained_performance is not None:  # Verificar si retained_performance es nulo
            cohesion = retained_performance / total_performance
            cohesion_factor = 1 + group['Temporadas_Juntas'].mean() * 0.1  # Factor de ponderación
            return cohesion * cohesion_factor
    return 0  # Devolver 0 si hay valores nulos o total_performance es 0


# Calcular la cohesión para la temporada siguiente (Equipo_prox, Temp_prox)
cohesion_weighted = player_data.groupby(['Equipo', 'Next_Temp']).apply(weighted_cohesion).reset_index(name='Cohesion_Weighted')
cohesion_performance = player_data.groupby(['Equipo', 'Next_Temp']).apply(performance_cohesion).reset_index(name='Cohesion_Performance')

# Renombrar las columnas para la fusión
cohesion_weighted = cohesion_weighted.rename(columns={'Next_Temp': 'Temp'})
cohesion_performance = cohesion_performance.rename(columns={'Next_Temp': 'Temp'})

# Crear 'cohesion_data'
cohesion_data = pd.merge(cohesion_weighted, cohesion_performance, on=['Equipo', 'Temp'], how='outer')

# Calcular 'Cohesion_Total' en 'cohesion_data'
weighted_weight = 0.6  # Peso para 'Cohesion_Weighted'
performance_weight = 0.4  # Peso para 'Cohesion_Performance'
cohesion_data['Cohesion'] = (cohesion_data['Cohesion_Weighted'] * weighted_weight) + (cohesion_data['Cohesion_Performance'] * performance_weight)

# Fusionar 'cohesion_data' con 'df_equipos'
df_equipos = pd.merge(df_equipos, cohesion_data[['Equipo', 'Temp', 'Cohesion']], on=['Equipo', 'Temp'], how='left')

# Asignar 0.5 de cohesión a equipos con 'Posicion_final_Ant' igual a 13
df_equipos.loc[df_equipos['Posicion_final_Ant'] == 13, 'Cohesion'] = 0.5

# Solución: Manejar valores nulos en "Cohesion_Total" después de la fusión
df_equipos['Cohesion'] = df_equipos['Cohesion'].fillna(0)  # Reemplazar NaN por 0
```

```python
# Función para calcular los puntos de una recta cualquiera.
def calcula_y_recta(X, m, b):
  y = m*X + b
  return y

# Función para separar aleatoriamente un dataframe en dos partes: entrenamiento y testeo, según una proporción de entrenamiento p
def SepararAleatoriamente(df, p):
  df['split'] = np.random.uniform(size=df.shape[0])
  train = df[df['split']<=p]
  test = df[df['split']>p]

  return train, test

# Función para graficar las variables X e y del dataframe df y superponer la recta de regresión y = mX + b
def GraficaVariablesYRecta(df, variable_X, variable_y, m, b):
  import matplotlib.pyplot as plt

  X = df[variable_X]
  y = df[variable_y]

  # Graficamos los puntos X, y
  plt.plot(X, y, 'o')

  # Añadimos la linea de regresión de test al gráfico
  X_recta = np.arange(0,X.max()+1,1)
  y_recta = calcula_y_recta(X_recta, m ,b)

  plt.plot(X_recta, y_recta)
  plt.show()

# Función para calcular nuestra métrica de error: el MSE
def MSE(
    X: np.ndarray,
    y: np.ndarray,
    m: float,
    b: float
) -> float:
    """
    función para calcular el mean squared error de la predicción de la recta
    respecto a los puntos (X,y).
    X: vector 1D, contiene las coordenadas X de los puntos.
    y: vector 1D, contiene las coordenadas y de los puntos.
    m: pendiente de la recta
    b: ordenada al orígen de la recta
    """
    assert X.size == y.size, 'la longitud de los vectores no coincide'
    y_recta = b + m * X
    mse = ((y - y_recta)**2).sum() / y.size
    return mse
```
